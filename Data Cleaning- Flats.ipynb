```python
import pandas as pd
import numpy as np
```

```python
flats = pd.read_csv("./flats.csv", encoding="latin-1")
flats.head()
```

```python
# dropping columns that aren't required
flats.drop(columns=["link","property_id"], inplace=True)
```

```python
flats.rename(columns={'area': "price_per_sqft"}, inplace=True)
```

```python
for i in flats["price"].unique():
    print(i)
```

```python
def fix_price(price):
    if price == "Price on Request":
        return np.nan
    elif pd.isna(price):
        return np.nan
    elif "Lac" in price:
        return float(price.split(" ")[0]) / 100
    else:
        return float(price.split(" ")[0])
    
flats["price"] = flats["price"].apply(fix_price)
```

```python
flats["bedRoom"] = flats["bedRoom"].str.split(" ").str.get(0).astype(float)
```
