```python
import pandas as pd
import numpy as np

pd.set_option("display.max_rows", None)
```

```python
houses = pd.read_csv("./houses.csv")
houses.head()
```

```python
#houses.duplicated().sum()

#dropping duplicates
houses.drop_duplicates(inplace=True)
```

```python
houses.columns
```

```python
# dropping unnecessary columns
houses.drop(columns=["link", "property_id"], inplace=True)
```

```python
houses.head()
```

```python
houses["price"].value_counts()
```

```python
houses["price"].isna().sum()
```

- There are 76 observations where price is null. Since we are building a price predictor, it doesn't make much sense to analyze data where the real price is not known.
- Also, it only accounts for around 7% of the total data, so it's safe to drop it.

```python
houses.drop(index= houses[houses["price"].isna()].index, inplace=True)
```

```python
houses.columns
```

```python
def fix_price(value):
    if "Crore" in value:
        return float(value.split(' ')[0])
    elif value == "Price on Request":
        return np.nan
    else:
        return round(float(value.split(' ')[0]) / 100, 2)
    
houses["price"] = houses["price"].apply(fix_price)
```

```python
houses.rename(columns={"rate":"price_per_sqft"}, inplace=True)
```

```python
houses["price_per_sqft"] = houses["price_per_sqft"].str.split(" ").str.get(1).str.split("/").str.get(0).str.replace(",","").astype(float)
```

```python
houses["bedRoom"] = houses["bedRoom"].str.split(" ").str.get(0).astype(int)
```

```python
houses.columns
```

```python
houses["bathroom"] = houses["bathroom"].str.split(" ").str.get(0).astype(int)
```

```python
houses["balcony"] = houses["balcony"].str.split(" ").str.get(0)
```

```python
houses["noOfFloor"] = houses["noOfFloor"].str.split(" ").str.get(0).astype(float)
```

```python
houses.rename(columns={"noOfFloor" : "floorNum"}, inplace=True)
```

### Calculating area from price and price per sqft

```python
houses["area"] = round((houses["price"]*10000000) / houses["price_per_sqft"])
```

### Inserting a new column property_type

```python
houses.insert(loc=1, column="property_type", value="house")
```
